```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
## Reading CSV
df = pd.read_csv('activities.csv',header=0)

## Adding End Date Column, using Start Date and duration
df['End Date'] = df['Start Date'] + df['Duration (in days)']

## Sorting by Maximum Finish Time
df=df.sort_values(by='End Date')
df.head(24)
```

```python
## plotting the activity vs time graph for a month
y=df['Quest'].to_list()

xmin = df['Start Date'].to_list()
xmax = df['End Date'].to_list()
rewards = df['Reward (in rupees)'].to_list()

#plt.plot(rewards)
plt.ylabel("Activities")
plt.xlabel("Time (days)")

plt.hlines(y, xmin, xmax)
plt.grid()
plt.show()
```

### Dynamic Programming approach

```python
class Job: 
    def __init__(self, activity, start, finish, reward):
        self.activity =  activity
        self.start  = int(start) 
        self.finish = int(finish) 
        self.reward  = reward 
  
  
# Finding the next compatible activity for Link using Binary Search
def binarySearch(job, start_index): 
  
    # Initialize 'lo' and 'hi' for Binary Search 
    lo = 0
    hi = start_index - 1
  
    # Perform binary Search iteratively 
    while lo <= hi: 
        mid = (lo + hi) // 2
        if job[mid].finish <= job[start_index].start: 
            if job[mid + 1].finish <= job[start_index].start: 
                lo = mid + 1
            else: 
                return mid 
        else: 
            hi = mid - 1
    return -1
  
# Returns the maximum possible 
# rewards from given array of activities
def schedule(activity): 
  
    # Create an array to store solutions of all possible activities.  table[i] 
    # stores the rewards for activities till arr[i] (including arr[i]) 
    n = len(activity)  
    table = [0 for _ in range(n)] 
  
    table[0] = activity[0].reward; 
      
    parent = [0]
    
    for i in range(1, n):
        # Find reward including the current activity
        activityRewards = activity[i].reward
        l = binarySearch(activity, i) 
        if (l != -1):
            activityRewards += table[l]
        
        if activityRewards > table[i - 1]:
            table[i] = activityRewards
            parent.append(1)
        else:
            table[i] = table[i - 1]
            parent.append(0)
        
    seq = []
    temp = 0
    i = len(parent) - 1
    while i >= 0:
        if parent[i] == 1:
            seq.append(activity[i].activity)
            temp = binarySearch(activity, i)
            i = temp
        else:
            i -= 1
    seq = seq[::-1]
    return "Maximized Earnings  is " + str(
        table[-1])+" rupees." + " One should choose the activities " + str(seq) + " in that order." 
```

```python
activities=[]
for i, reward in enumerate(rewards):
    activities.append(Job(y[i], xmin[i],xmax[i],reward))

schedule(activities)
```

### The above solution works for any given activity board as it takes into account the maximum earnings from those activities.

#### Step 1. 
We first sort all our activities based on their finish times and compare the activities one by one to find the next compatible activity because activities cannot overlap each other.

#### Step 2.
We use a binary search to find the most compatible activity. This takes us O(lg n) time.

#### Step 3.
We do this n times and while we do this we save the activity names to out put all the possible activities in the end.

#### Step 4. 
Keep adding the reward of all the compatible, highest rewards and display the sum in the end.
